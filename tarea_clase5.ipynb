```python
# 1.Calculadora Simple: Crea un programa que solicite al usuario dos números y una operación 
# (suma, resta, multiplicación o división). Luego, utilizando estructuras condicionales (if, elif, else), 
# realiza la operación seleccionada e imprime el resultado.

num1=int(input("Ingrese primer numero"))
num2=int(input("Ingrese segundo numero"))
operacion=input("Seleccione una se las siguientes operaciones: suma, resta, multiplicacion o division")
if operacion == "suma":
    suma=num1+num2
    print("El resultado de la suma es:",suma)
elif operacion == "resta":
    resta=num1-num2
    print("El resultado de la resta es:",resta)
elif operacion == "multiplicacion":
    multiplicacion=num1*num2
    print("El resultado de la multiplicación es: ",multiplicacion)
elif operacion == "division":
    if num2 != 0:
        division=num1/num2
        print("El resultado de la division es:",division)
```

```python
# 2. Contador de Números Pares e Impares: Escribe un programa que recorra una lista de números del 1 al 20.
# Utilizando un bucle for, determina cuántos de estos números son pares y cuántos son impares. Al final, 
# imprime la cantidad de números pares e impares.
contador_pares=0
contador_impares=0

for i in range(1,21):
    if i % 2 == 0:
        contador_pares += 1
    else:
        contador_impares += 1
print("La cantidad de numeros pares es: ", contador_pares)
print("La cantidad de numeros pares es: ", contador_impares)
  

```

```python
# 3.Validación de Contraseña: Pide al usuario que ingrese una contraseña.
# Utiliza un bucle while para seguir solicitando la contraseña hasta que coincida con una contraseña predefinida 
# (por ejemplo, "python123"). Muestra un mensaje de éxito cuando la contraseña es correcta.
contraseña_correcta="python123"
contraseña=input("Ingrese la contraseña")
while contraseña!=contraseña_correcta:
    print("Contraseña incorrecta")
    contraseña=input("Ingrese la contraseña nuevamente")
print("Contraseña correcta")
```
